```python
dic_columns = {
    'LIMIT_BAL' : 'limite',
    'CHECKING_ACCOUNT' : 'Cuenta_Corriente',
    'EDUCATION' : 'escolaridad',
    'MARRIAGE' : 'estado_civil',
    'AGE' : 'edad',
       'BILL_AMT' : 'valor_factura',
    'PAY_AMT' : 'valor_pago',
    'DEFAULT' : 'moroso'
}
```

#Test De Hipotesis

```python
sns.lmplot(x='iu',y='valor_factura',data=tarjetas, col='moroso')
```

```python
from scipy.stats import ranksums
```

H<sup>null<sup>

La distribucion de los grupos moroso y no moroso es la misma

Hnull

La distribucion de los grupos moroso y no morroso no es la misma

```python
moroso = tarjetas.query("moroso == 1").valor_factura
moroso
```

```python
no_moroso = tarjetas.query("moroso == 0").valor_factura
no_moroso
```

```python
resultado_residual = ranksums(moroso, no_moroso)
```

```python
print(f"El p_value de nuestro test es {resultado_residual.pvalue}")
```
